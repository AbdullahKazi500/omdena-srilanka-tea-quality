```python

from glob import glob
import matplotlib.pyplot as plt
from augmentation import augment_every_image, multiple_augmentations_on_an_image
```

```python
image_path = ['./P088.jpg']
```

```python
augment_every_image(images=image_path, save_path='./augment_every_image_result', 
                    resize_shape=(256, 256), equalize=False)
```

```python
# Helper functions

def load_data(path):
    """ Return the image files in a given part """
    image_paths = sorted(glob(path+'/*.jpg'))
    print('Number of images returned {}'.format(len(image_paths)))
    return image_paths

```

```python
img_paths = load_data('./augment_every_image_result')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(15, 10), subplot_kw={'xticks': [], 'yticks': []})

for ax, image in zip(axes.flat, img_paths):
    ax.imshow(plt.imread(image))
    ax.set_title(image.split('/')[-1].split('.')[0])
    
plt.show()
```

```python
image_path = ['./P012.jpg']
multiple_augmentations_on_an_image(images=image_path, 
                                   save_path='./multiple_augmentations', 
                                   resize_shape=(256, 256))
```

```python
img_paths = load_data('./multiple_augmentations')

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10), subplot_kw={'xticks': [], 'yticks': []})

for ax, image in zip(axes.flat, img_paths):
    ax.imshow(plt.imread(image))
    ax.set_title(image.split('/')[-1].split('.')[0])
    
plt.show()
```
